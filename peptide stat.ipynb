```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stat_plot_all(length):
    data = pd.read_csv('Labeled-Filled-HLA-A_0201.csv')
    data_0 = data.loc[data['peptide_length'] == length]
    position = np.array(data_0['sequence'])
    binding = np.array(data_0['label'])
    P = 0
    N = 0
    for i in range(len(position)):
        if binding[i] == 'P':
            P += 1
        else:
            N += 1
    r3 = ['P', 'N']
    bars3 = [P/(P + N), N/(P + N)]
    plt.figure(figsize=(5,5))
    barWidth = 0.25
    plt.bar(r3[0], bars3[0], color='g', width=barWidth, edgecolor='white', label = 'Positive')
    plt.bar(r3[1], bars3[1], color='b', width=barWidth, edgecolor='white', label = 'Negative')
    #Create legend & Show graphic
    plt.title('Length = %d'%length, fontweight = 'bold')
    plt.tight_layout()
    plt.legend(loc = 'upper center')
    plt.show()
    return
```

```python
stat_plot_all(8),stat_plot_all(9),stat_plot_all(10),stat_plot_all(11),stat_plot_all(12)
```

```python
def stat_plot(i):
    data = pd.read_csv('Labeled-Filled-HLA-A_0201.csv')
    data_9 = data.loc[data['peptide_length'] == 9]
    position = np.array(data_9['sequence'])
    binding = np.array(data_9['label'])
    P = 0
    N = 0
    count = {
        'I': [0, 0, 0],
        'V': [0, 0, 0],
        'L': [0, 0, 0],
        'F': [0, 0, 0],
        'C': [0, 0, 0],
        'M': [0, 0, 0],
        'A': [0, 0, 0],
        'G': [0, 0, 0],
        'T': [0, 0, 0],
        'S': [0, 0, 0],
        'W': [0, 0, 0],
        'Y': [0, 0, 0],
        'P': [0, 0, 0],
        'H': [0, 0, 0],
        'E': [0, 0, 0],
        'Q': [0, 0, 0],
        'D': [0, 0, 0],
        'N': [0, 0, 0],
        'K': [0, 0, 0],
        'R': [0, 0, 0]
    }
    if i == 4 or i ==  5 or i == 6:
        print('NAN')
        return
    else:
        for j in range(len(position)):
            count[position[j][i]][0] += 1
            if binding[j] == 'P':
                count[position[j][i]][1] += 1
                P = P + 1
            else: 
                count[position[j][i]][2] += 1
                N = N + 1
    # set width of bar
    plt.figure(figsize=(15,5))
    barWidth = 0.25

    # set height of bar
    bars1 = [elem[1] for elem in count.values()]
    bars2 = [elem[2] for elem in count.values()]
    bars1_nor = [num1/P for num1 in bars1]
    bars2_nor = [num2/N for num2 in bars2]
    # Set position of bar on X axis
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]

    # Make the plot
    #plt.subplot(121)
    plt.bar(r1, bars1_nor, color='g', width=barWidth, edgecolor='white', label = 'Positive')
    plt.bar(r2, bars2_nor, color='b', width=barWidth, edgecolor='white', label = 'Negative')

    # Add xticks on the middle of the group bars
    if i < 4:
        plt.xlabel('Position %d'%(i + 1), fontweight='bold')
    else:
        plt.xlabel('Position %d'%(i - 2), fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], list(count.keys()))
    plt.legend()
    
    #plt.subplot(122)
    #r3 = ['P', 'N']
    #bars3 = [P/(P + N), N/(P + N)]
    #plt.bar(r3[0], bars3[0], color='g', width=barWidth, edgecolor='white', label = 'Positive')
    #plt.bar(r3[1], bars3[1], color='b', width=barWidth, edgecolor='white', label = 'Negative')
    # Create legend & Show graphic
    #plt.legend()
    plt.show()
    return
```
